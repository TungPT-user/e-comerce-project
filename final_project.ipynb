```python
import pandas as pd

df = pd.read_csv('/content/Amazon Sale Report.csv')
df.head()
```

```python
df = df.rename(columns={'Order ID' : 'Order_ID'})
df = df.rename(columns={'Courier Status' : 'Courier_Status'})
df = df.rename(columns={'ship-service-level' : 'ship_service_level'})
df = df.rename(columns={'ship-city' : 'ship_city'})
df = df.rename(columns={'ship-state' : 'ship_state'})
df = df.rename(columns={'ship-country' : 'ship_country'})
df.columns = df.columns.str.strip()  # Xóa khoảng trắng thừa
df = df.rename(columns={"Sales Channel": "Sales_Channel"})
df
```

```python
# check dòng, cột
df.shape
```

```python
#check thông tin
df.info()
```

```python
# thấy Unnamed: 22 và fulfilled-by bị null > 40% => drop
# Cột currency,style không cần thiết trong việc phân tích => drop
# đây là bài phân tích dữ liệu doanh số Baseline Sales nên không cần promotion ids nên => drop
# ship-postal-code không có giúp ích cho bài phân tích => drop
# tương tự với ASIN và SKU -- cột 8 và 11
# check ship_country thấy chỉ có mỗi IN => drop
df = df.drop(['currency'], axis=1)
df = df.drop(['Style'], axis=1)
df = df.iloc[:, :-2]
df = df.drop(['ship-postal-code'], axis=1)
df = df.drop(['promotion-ids'], axis=1)
df = df.drop(['SKU'], axis=1)
df = df.drop(['ASIN'], axis=1)
df['ship_country'].unique()
df = df.drop(['ship_country'], axis=1)
df
```

```python
# check giá trị trùng lặp
df.duplicated().sum()
```

```python
df.describe()
```

```python
# Chia thành 3 Status Shipped, Pending, Cancelled

df.loc[df["Status"] == "Shipped - Delivered to Buyer", "Status"] = "Shipped"
df.loc[df["Status"] == "Shipped - Returned to Seller", "Status"] = "Shipped"
df.loc[df["Status"] == "Shipped - Picked Up", "Status"] = "Shipped"
df.loc[df["Status"] == "Pending - Waiting for Pick Up", "Status"] = "Pending"
df
```

```python
# thay tất cả các giá trị NaN = Shipped
df.loc[df["Courier_Status"].isna(), "Courier_Status"] = "Shipped"
df
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df
```

```python
df["Sales_Channel"] = df["Sales_Channel"].replace("Amazon.in", "Amazon")
df
```

```python
df['ship_city'] = df['ship_city'].str.upper()
df['ship_state'] = df['ship_state'].str.upper()
df
```

```python
df.to_csv('Amazon-sale-report-clean-4.csv', index=False)
```
